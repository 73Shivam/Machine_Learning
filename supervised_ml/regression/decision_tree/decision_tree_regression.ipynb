```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df =  pd.read_csv('datasets/oil_dataset.csv')
df.head()
```

```python
df.info()
```

```python
fig = px.scatter_3d(df,x='Depth',y='Mendacium',z='Price',size='Price',color='Mendacium',)
fig.show()
```

```python
X = df[['Mendacium','Depth']]
y = df['Price']
scaler = StandardScaler()
x = scaler.fit_transform(X)
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(x,y, test_size=.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
reg = DecisionTreeRegressor(max_depth=15)
reg.fit(xtrain, ytrain)
reg.score(xtest, ytest) * 100
```

```python
reg = LinearRegression()
reg.fit(xtrain, ytrain)
reg.score(xtest, ytest) * 100
```

```python
pf = PolynomialFeatures()
xtrain_p = pf.fit_transform(xtrain)
xtest_p  = pf.transform(xtest)
reg = LinearRegression()
reg.fit(xtrain_p, ytrain)
reg.score(xtest_p, ytest) * 100
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor(n_estimators=10)
reg.fit(xtrain, ytrain)
reg.score(xtest, ytest) * 100
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
x = df[['MOMENT','LOCATION','CRIME','VICTIM_SEX','VICTIM_AGE']]
y= df['OCCURRENCES']
```

```python
moment_enc = OneHotEncoder()
moments = moment_enc.fit_transform(df[['MOMENT']]) .toarray()
moments.astype(int)
```

```python
location_enc = OneHotEncoder()
locations = location_enc.fit_transform(df[['LOCATION']]) .toarray()
locations.astype(int)
```

```python
crime_enc = OneHotEncoder()
crimes = crime_enc.fit_transform(df[['CRIME']]) .toarray()
crimes.astype(int)
```

```python
gender_labeler = LabelEncoder() 
gender = gender_labeler.fit_transform(df[['VICTIM_SEX']])
gender
```

```python
x = np.concatenate([ 
                    df[['VICTIM_AGE']],  # dataframe
                     gender.reshape(-1,1), # reshape 2d array
                     crimes[:,:-1] ,  # removed 1 last col
                     locations[:,:-1], # removed 1 last col
                     moments[:,:-1] # removed 1 last col
                    ], axis=1)
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(x,y, test_size=.2, random_state=0)
```
