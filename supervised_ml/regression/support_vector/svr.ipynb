```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
df = pd.read_csv('datasets/Position_Salaries.csv')
```

```python
df.head()
```

```python
sns.regplot(x='Level',y='Salary',data=df,)
```

```python
X =df[['Level']]  # vector or 2d array or dataframe
Y = df['Salary']  # scaler or 1d array or series
sx = StandardScaler() # scale object
sy = StandardScaler() # scale the y column also in case of svr
x = sx.fit_transform(X) # converted the data
# [[0,1,2,3,4,5]] 1 row in a 2d array
# [[0],[1],[2],[3],[4],[5]] # 1 col in a 2d array
y = sy.fit_transform(Y.values.reshape(-1,1)) # y series > numpy values > rows to columns using reshape
```

```python
reg = SVR()
reg.fit(x,y)
reg.score(x,y) * 100
```

```python
ypred = reg.predict(x)
ypred = sy.inverse_transform(ypred)
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,ypred,color='blue')
```
