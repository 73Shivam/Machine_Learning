```python
import pandas as pd
import numpy as np
import plotly.express as px
import os
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataset ="Salary_Data.csv"  
if os.path.exists(dataset):
    print('file found')
else:
    print("check path of file")
```

```python
df = pd.read_csv(dataset)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df[['YearsExperience']] # x data that should be taken as a vector or 2D or dataframe
y = df['Salary'] # y data is taken as scaler or 1D or series
```

```python
xtrain, xtest, ytrain, ytest =  train_test_split(X,y,test_size= .2)
```

```python
reg = LinearRegression()
reg.fit(xtrain, ytrain) # here the algorithm tries to understand the data
```

```python
reg.score(xtest,ytest) * 100
```

```python
ypred = reg.predict(xtest)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(xtest,ytest,'or',label='original')
plt.plot(xtest,ypred,'-g',label='prediction')
plt.legend=True
plt.show()
```

```python
plt.plot(xtrain,ytrain,'or',label='original')
plt.plot(xtrain,reg.predict(xtrain),'-g',label='prediction')
plt.legend=True
plt.show()
```
